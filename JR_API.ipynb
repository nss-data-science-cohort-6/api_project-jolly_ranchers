```python
endpoint_pop= f'https://api.census.gov/data/2020/acs/acs5?get=NAME,B01001_001E&for=tract:*&in=county:037&in=state:47'

response = requests.get(endpoint_pop)

response.text
```

```python
res = response.json()
res
```

```python
acs_pop = pd.DataFrame(res)
acs_pop
```

```python
header = acs_pop.iloc[0] 
acs_pop = acs_pop[1:] 
acs_pop.columns = header
```

```python
acs_pop = acs_pop.rename(columns={'B01001_001E':'Population'})
```

```python
acs_pop
```

```python
endpoint_medinc = 'https://api.census.gov/data/2020/acs/acs5/subject?get=NAME,S1901_C01_012E&for=tract:*&in=state:47&in=county:037'

response = requests.get(endpoint_medinc)

response.text
```

```python
res = response.json()
res
```

```python
acs_medinc = pd.DataFrame(res)
acs_medinc
```

```python
header = acs_medinc.iloc[0] 
acs_medinc = acs_medinc[1:] 
acs_medinc.columns = header
```

```python
acs_medinc = acs_medinc.rename(columns={'S1901_C01_012E':'Median Income'})
```

```python
acs_medinc
```

```python
acs_pop_medinc = pd.merge(acs_pop, acs_medinc, on=['NAME', 'state', 'county', 'tract'])
acs_pop_medinc
```

```python
type(acs_pop_medinc)
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
from folium.plugins import FastMarkerCluster
```

```python
agg_burg['geometry'] = gpd.points_from_xy(agg_burg['longitude'], agg_burg['latitude'])

agg_burg.head()
```

```python
tn_tracts = gpd.read_file('./tn tracts/')
```

```python
type(tn_tracts)
```

```python
aggburg_tntracts = gpd.GeoDataFrame(agg_burg, 
                           crs = tn_tracts.crs, 
                           geometry = agg_burg['geometry'])
```

```python
aggburg_tntracts
```

```python
aggburg_tntracts.columns
```

```python
aggburg_by_censustract = gpd.sjoin(aggburg_tntracts, tn_tracts, predicate = 'within')
```

```python
aggburg_by_censustract
```

```python
aggburg_by_censustract['incident_number'].value_counts()
```

```python
aggburg_by_censustract = aggburg_by_censustract.drop_duplicates('incident_number')
```

```python
aggburg_by_censustract.rename(columns = {'TRACTCE':'tract'}, inplace = True)
aggburg_by_censustract
```

```python
burglaries = aggburg_by_censustract['tract'].value_counts().reset_index(name='count_burglary')
burglaries = burglaries.rename(columns = {'index':"tract"})
burglaries
```

```python
# Census tract 16300 had the highest number of aggravated burglaries at 31.
```

```python
merged_question_5 = pd.merge(left=acs_pop_medinc, right=burglaries, on=['tract'], how='outer')
merged_question_5
```

```python
merged_question_5.columns
```

```python
merged_question_5 = merged_question_5.astype({'Population': int})
merged_question_5 = merged_question_5.astype({'Median Income': int})
```

```python
question5_filtered = merged_question_5[merged_question_5['Population'] > 0]
question5_filtered = merged_question_5[merged_question_5['Median Income'] > 0]
```

```python
question5_filtered.head()
```

```python
question5_filtered.shape
```

```python
question5_filtered['burglaries/thousand'] = 1000*question5_filtered['count_burglary']/question5_filtered['Population']
question5_filtered
```

```python
question5_filtered.to_csv('api.csv')
```
